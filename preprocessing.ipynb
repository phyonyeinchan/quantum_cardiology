```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

print("Libraries imported successfully!")
```

```python
# UCI Repository ရှိ Heart Failure Clinical Records ရဲ့ CSV ဖိုင် Link အစစ် ဖြစ်ပါတယ်
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00519/heart_failure_clinical_records_dataset.csv"
df = pd.read_csv(url)

# ဒေတာရဲ့ အရွယ်အစား (လူနာ အရေအတွက်နှင့် အချက်အလက် အရေအတွက်) ကို စစ်ဆေးခြင်း
print(f"Dataset Shape: {df.shape}")

# ဒေတာ၏ ပထမဆုံး ၅ ကြောင်းကို ဇယားကွက်ဖြင့် နမူနာ ကြည့်ရှုခြင်း
df.head()
```

```python
# ကူးယူရမည့် ကုဒ်အပြည့်အစုံ ဖြစ်ပါတယ်
# ပြောင်းလဲရှုပ်ထွေးစေမည့်အချက်များ (Confounders) ကို ရွေးထုတ်ခြင်း
X_features = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 
              'ejection_fraction', 'platelets', 'serum_creatinine', 
              'serum_sodium', 'sex', 'smoking']

X = df[X_features]
w = df['high_blood_pressure'] # Treatment (သွေးတိုးရှိ/မရှိ)
y = df['DEATH_EVENT']         # Outcome (သေဆုံးမှု ရှိ/မရှိ)

print("Variables assigned successfully!")
print(f"Features (X) Shape: {X.shape}")
print(f"Treatment (w) Count: {w.shape[0]}")
print(f"Outcome (y) Count: {y.shape[0]}")
```

```python
# ကုဒ်အပြည့်အစုံကို ကူးယူပါ
# ဒေတာများကို ကွန်ပျူတာသင်ယူရန် (Train 80%) နှင့် စမ်းသပ်ရန် (Test 20%) ခွဲခြားခြင်း
X_train, X_test, w_train, w_test, y_train, y_test = train_test_split(
    X, w, y, test_size=0.2, random_state=42
)

# Quantum Computing စနစ်၏ သဘာဝအရ Data များကို Scale ညှိပေးရန် လိုအပ်ပါသည်
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Data Scaling & Train/Test split completed successfully!")
print(f"Training Features Size: {X_train_scaled.shape}")
print(f"Testing Features Size: {X_test_scaled.shape}")
```
