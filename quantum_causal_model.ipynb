```python
from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes

# မိမိဒေတာ၏ Feature အရေအတွက်ပေါ် မူတည်ပြီး num_qubits ကို ပြောင်းလဲနိုင်သည် (ဥပမာ- ၄ ခု)
num_qubits = 4 

# Feature Map နှင့် Ansatz ကို ကြေညာခြင်း
feature_map = ZZFeatureMap(feature_num_qubits=num_qubits, reps=2)
ansatz = RealAmplitudes(num_qubits=num_qubits, reps=2)
```

```python
# ၁။ လိုအပ်သော Libraries အားလုံးကို တစ်ခါတည်း တစ်စုတစ်စည်းတည်း ခေါ်ယူခြင်း
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from qiskit import QuantumCircuit
from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes
from qiskit_machine_learning.algorithms import VQC
from qiskit_algorithms.optimizers import COBYLA

print("Step 1: Libraries Loaded.")

# ၂။ ဆေးဘက်ဆိုင်ရာ Cardiology Data ကို ဆွဲယူပြီး အသင့်ပြင်ဆင်ခြင်း
url = "https://uci.edu"
df = pd.read_csv(url)

X_features = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 
              'ejection_fraction', 'platelets', 'serum_creatinine', 
              'serum_sodium', 'sex', 'smoking']

X = df[X_features]
w = df['high_blood_pressure'] # Treatment
y = df['DEATH_EVENT']         # Outcome

# Data Scaling ညှိခြင်း
X_train, X_test, w_train, w_test, y_train, y_test = train_test_split(X, w, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

print("Step 2: Cardiology Data Preprocessed.")

# ၃။ Quantum Circuit ဆောက်ခြင်း (10 Qubits သတ်မှတ်ခြင်း)
num_qubits = X_train_scaled.shape[1] # 10 Features ရှိ၍ 10 ရပါမည်

# အမှားကင်းစင်သော အဓိက ကန့်သတ်ချက်များ
feature_map = ZZFeatureMap(feature_dimension=num_qubits, reps=1, entanglement='linear')
ansatz = RealAmplitudes(num_qubits=num_qubits, reps=1)

print(f"Step 3: Quantum Circuit created with {num_qubits} Qubits.")

# ၄။ Variational Quantum Classifier (VQC) စတင် Train ခြင်း
optimizer = COBYLA(maxiter=20) # သင့် Mac M4 ပေါ်တွင် စက္ကန့်ပိုင်းအတွင်း မြန်မြန်ပြီးစေရန် maxiter ကို ၂၀ ဟု သတ်မှတ်ပါသည်
vqc = VQC(
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer
)

print("Step 4: Training Quantum Model (VQC)... Please wait a moment...")
vqc.fit(X_train_scaled, w_train.values)

print("🎉 Congratulations! Quantum Model Training Completed Successfully!")
```

```python
# ၁။ လိုအပ်သော Libraries အားလုံးကို တစ်ခါတည်း တစ်စုတစ်စည်းတည်း ခေါ်ယူခြင်း
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from qiskit import QuantumCircuit
from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes
from qiskit_machine_learning.algorithms import VQC
from qiskit_algorithms.optimizers import COBYLA

print("Step 1: Libraries Loaded.")

# ၂။ ဆေးဘက်ဆိုင်ရာ Cardiology Data ကို ဆွဲယူခြင်း (လင့်ခ်အမှန်ကို သေချာစွာ စစ်ဆေးပါ)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00519/heart_failure_clinical_records_dataset.csv"
df = pd.read_csv(url)

X_features = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 
              'ejection_fraction', 'platelets', 'serum_creatinine', 
              'serum_sodium', 'sex', 'smoking']

X = df[X_features]
w = df['high_blood_pressure'] # Treatment
y = df['DEATH_EVENT']         # Outcome

# Data Scaling ညှိခြင်း
X_train, X_test, w_train, w_test, y_train, y_test = train_test_split(X, w, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

print("Step 2: Cardiology Data Preprocessed.")

# ၃။ Quantum Circuit ဆောက်ခြင်း (10 Qubits သတ်မှတ်ခြင်း)
num_qubits = X_train_scaled.shape[1] # Features ၁၀ ခုရှိ၍ ၁၀ ရပါမည်
feature_map = ZZFeatureMap(feature_dimension=num_qubits, reps=1, entanglement='linear')
ansatz = RealAmplitudes(num_qubits=num_qubits, reps=1)

print(f"Step 3: Quantum Circuit created with {num_qubits} Qubits.")

# ၄။ Variational Quantum Classifier (VQC) စတင် Train ခြင်း
optimizer = COBYLA(maxiter=20) # အချိန်သက်သာစေရန် Iteration အနည်းငယ်ဖြင့် စမ်းပါမည်
vqc = VQC(
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer
)

print("Step 4: Training Quantum Model (VQC)... Please wait a moment...")
vqc.fit(X_train_scaled, w_train.values)

print("🎉 Congratulations! Quantum Model Training Completed Successfully!")
```
